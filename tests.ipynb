```python
def run_chunk_stage(embedding_args: AttrDict) -> None:
    all_passages = []
    print("Stage 1/2: chunking with src.data.fast_load_jsonl_shard")
    for shard_id in embedding_args.shard_ids:
        passages = fast_load_jsonl_shard(embedding_args, shard_id)
        num_passages = 0 if passages is None else len(passages)
        print(f"  shard={shard_id}: {num_passages} chunked passages")
        all_passages.append(passages)
    return all_passages
```

```python
all_passages = run_chunk_stage(embedding_args)
print(type(all_passages))
```

```python
import textwrap

def pretty_print_dict(text, width=70):
    """
    Pretty print a dictionary with string values wrapped to a given width.
    
    :param data: Dictionary with string values
    :param width: Maximum line width before wrapping
    """
    wrapped = textwrap.fill(text, width=width)
    indented = textwrap.indent(wrapped, prefix="    ")
    print(indented)
    print()  # blank line between entries
```
